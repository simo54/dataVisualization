With the following data visualization, we are going to display 2 graphs related to the famous Harry Potter Saga.

We compare in this file the first and the last movie ('Harry Potter and the Philosopher's Stone' and 'Harry Potter and the Deathly Hallows – Part 2').

The last chapter of the saga has been divided into 2 movies but for this research, we will consider 'Harry Potter and the Deathly Hallows – Part 2' as a stand alone movie.

Let's point out a few information:
- Hp1 was released in November (2001). 
- Hp8 was released in July (2011).
- We will not consider marketing or movie budget. 
- Data Source: Box Office Mojo ("harrypotter.csv")

```python
import pandas as pandas
```

```python
from matplotlib import pyplot as plt
```

```python
# Getting data from file source
data=pandas.read_csv('harrypotter.csv')

# Declaring variables
hp1 = data[data.film == 1]
hp8 = data[data.film == 8]

# Totatel revenue (sum of all movies)
totalRevenueAllMovies = []
for i in range(1,9):
    totalRevenueAllMovies.append(data[data.film == i]['revenue'].sum())

# Declring variables for pie display
hp1TotalRevenue=totalRevenueAllMovies[0]
hp2TotalRevenue=totalRevenueAllMovies[1]
hp3TotalRevenue=totalRevenueAllMovies[2]
hp4TotalRevenue=totalRevenueAllMovies[3]
hp5TotalRevenue=totalRevenueAllMovies[4]
hp6TotalRevenue=totalRevenueAllMovies[5]
hp7TotalRevenue=totalRevenueAllMovies[6]
hp8TotalRevenue=totalRevenueAllMovies[7]
```

```python
# Drawing the line chart 
plt.plot(hp1.weeknum, hp1.revenue / 10**6)
plt.plot(hp8.weeknum, hp8.revenue / 10**6)
plt.legend(['Harry Potter: 1', 'Harry Potter: 8' ])
plt.xlabel('Weeks on theaters')
plt.ylabel('Revenue(Mln$)')
plt.title('View of Revenue during weeks')
plt.show()
```

Observing the line chart, we can see how the movies performed during their projection. The HP1 has been in theaters six weeks more than HP8. But the impact on the launch of HP8 was extremely considerable.

```python
# Showing in a pie the individual total revenue of each movie
labels = 'HP1', 'HP2', 'HP3', 'HP4', 'HP5', 'HP6', 'HP7', 'HP8'
sizes = [hp1TotalRevenue,hp2TotalRevenue,hp3TotalRevenue,hp4TotalRevenue,hp5TotalRevenue,hp6TotalRevenue,hp7TotalRevenue,hp8TotalRevenue]
explode=(0.3,0,0.2,0,0,0,0,0.3)
fig1, ax1 = plt.subplots()
ax1.set_title('% Impact of each movie')
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90, textprops={'fontsize': 12 })
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Thanks to the pie graph we can observe that the first and last movies are the films that made more revenue in these past years. It would be interesting to gather more details on the HP3 and the reason why this chapter has not been as successful as the others.
